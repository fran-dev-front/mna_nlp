#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el result de los primeros 10 comentarios resultantes. 
   




```python
limpieza = []
for comment in docs:
    paso1 = re.sub(r'\n', "", comment)
    paso2 =  re.sub(r'[A-Z]', lambda x: x.group().lower(), paso1)
    paso3 = re.sub(r'\s{2,}', ' ', paso2)
    limpieza.append(re.sub(r'[^\w\s]', "", paso3))
```

```python
limpieza[0:10]
```

*   **Pregunta 14.**  

Con el result de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus. 

Es decir, al final de este proceso de tokenización, debes tener como result una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus. 

```python
tokenizacion = []
for comment in limpieza:
    tokenizacion.append(re.findall(r"\w*(?:\w+)", comment))
tokenizacion
```

```python
tokens = 0
for frase in tokenizacion:
    #print(frase)
    tokens+=len(frase)
tokens
        
```
